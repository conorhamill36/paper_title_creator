```python
from datetime import datetime
```

```python
def find_date(text):
    doc = nlp(text)
    print([ent.label_ for ent in doc.ents])
    dates = [ent for ent in doc.ents if ent.label_ == 'DATE']
    print(dates)
    years = list(filter(lambda x: x.isnumeric(), (str(date) for date in dates)))
    print(years)
    current_year = datetime.now().year
    if not years:
        return '????'
    return str(max([int(year) for year in years if int(year) <= current_year]))
find_date(text)
```

```python
def find_name(text):
    doc = nlp(text)
    labels = [ent.label_ for ent in doc.ents]
    persons = [ent for ent in doc.ents if ent.label_ =='PERSON']
    
    # Getting first person mentioned in paper
    person = persons[0].text
    # print(person)
    # remove number from string
    person = ''.join([char for char in person if char.isalpha() or char.isspace()])
    
    # split on spaces
    person_split = person.split(' ')
    
    # Getting final full name out of split
    surname = [split for split in person_split if len(split) > 1][-1]
    
    return surname
find_name(text)
```

```python
from collections import Counter
```

```python
def find_subject(text):
    # TODO: remove singular and plural words
    doc = nlp(text)
    words = [token.text.lower() for token in doc if token.pos_ == 'NOUN']
    word_freq = Counter(words)
    most_freq_words = [freq_word[0] for freq_word in word_freq.most_common(3)]
    return most_freq_words
subjects = find_subject(text)
```

```python
def get_paper_name(text, file_extension='.pdf'):
    year = find_date(text)
    surname = find_name(text)
    subject_list = find_subject(text)
    
    subject_slug = '_'.join(subject_list)
    paper_name = '_'.join([year, surname, subject_slug]) + file_extension 
    return paper_name
get_paper_name(text)
```
